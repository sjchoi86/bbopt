### Select a subset using k-DPP

```python
xmin,xmax,xdim = 0,5,2
x_minmax = np.tile(np.array([[xmin,xmax]]),(xdim,1))
```

```python
n_raw,n_sel,invlen = 5000,100,1000
x_sub_concat = np.zeros(shape=(0,2))
for i_idx in range(1):
    x_sub = get_x_sub_kdpp(x_minmax,n_sel,n_raw=n_raw,invlen=invlen)
    x_sub_concat = np.concatenate((x_sub_concat,x_sub),axis=0)
    
# Plot k-DPP subset 
plot_scatter(x=x,c=c,
             x2=x_sub,fmt2='o',col2='k',mew2=2,ms2=10,label2='k-DPP(%d)'%(n_sel))
```
